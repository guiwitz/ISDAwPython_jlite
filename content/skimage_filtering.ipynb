# Using scikit-image in Jupyterlite

```python
import skimage
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
myimage = np.random.randint(0,255,(20,20))
```

```python
plt.imshow(myimage)
plt.show()
```

```python
my_image_filtered = skimage.filters.gaussian(myimage)
```

```python
plt.imshow(my_image_filtered)
plt.show()
```
